```python
x = [1,2,3,4,5]
y1= [9,5,6,7,8]
y2= [9,8,12,7,8]


plt.plot(x,y1, color = "red", marker =  "o", linestyle = '--', linewidth = 3, markersize = 7,label = 'line y1')
plt.plot(x,y2, color = "b", marker =  "o", linestyle = '--', linewidth = 3, markersize = 10, label = 'line y2')

plt.xlabel("x data axis")
plt.xlabel("y data axis")

plt.title("line to understand")


plt.show()
```
